```python
# ko
import easyocr # EasyOCR 라이브러리
import cv2 # openCV2 라이브러리
import matplotlib.pyplot as plt # matplotlib.pyplot 라이브러리.함수
```

```python
# 읽어올 언어 설정

reader = easyocr.Reader(['ko', 'en']) 
```

```python
# openCV로 image 읽기

img_path = 'easy_1.png' # 이미지 경로

img = cv2.imread(img_path) # 이미지 읽기

plt.figure(figsize=(8, 8)) # 사이즈는?

plt.imshow(img[:, :, ::-1]) # bgr(openCV 이미지 표현방식)에서 rgb(삼원색 표현) 채널로 변경

plt.axis('off') # 부분적으로 제거 하고 싶을때 사용(off)

plt.show() # 출력!
```

```python
# image 결과물을 text로 출력

result = reader.readtext(img_path) # 결과 = readtext함수()

result # 결과물 확인 
```

```python
THRESHOLD = 0.5 # threshold : 기준값에서 0.5 숫자보다 작으면 0, 크면 1로 한다는 의미이다.(반올림)

for bbox, text, conf in result: 
  if conf > THRESHOLD: # conf가 threshold보다 크면
    print(text) # text로 출력
    cv2.rectangle(img, pt1=bbox[0], pt2=bbox[2], color=(0, 255, 0), thickness=3) 
    # img : 그림을 그릴 이미지 파일
    # start(시작좌표 == pt1) / end(종료좌표 == pt2) 
    # cv2(openCV).그리기함수(사각형), 이미지,pt1은 왼쪽위 좌표
    # pt2는 오른쪽 아래 좌표
    # rgb 형태의 color
    # thickness :  선의 두께

    plt.figure(figsize=(8,8))

    plt.imshow(img[:, :, ::-1])

    plt.axis('off')

    plt.show()
```

```python
# 함수화

THERESHOLD = 0.5

reader = easyocr.Reader(['ko', 'en'])

def read(img_path):
  
  img = cv2.imread(img_path)

  result = reader.readtext(img_path)

  r = []

  for bbox, text, conf in result:
    if conf > THRESHOLD:
      r.append(text)
      cv2.rectangle(img, pt1=(int(bbox[0][0]), int(bbox[0][1])),pt2=(int(bbox[2][0]), int(bbox[2][1])), color=(0, 255, 0), thickness=2)
  
  print(r)

  plt.figure(figsize=(8, 8))

  plt.imshow(img[:, :, ::-1])

  plt.axis('off')

  plt.show()
```
